## Basic text preprocessing (future step)

### Subtask:
Clean the text data by removing noise like URLs, mentions, hashtags, and special characters. Convert text to lowercase and potentially perform tokenization.


**Reasoning**:
Define a function to clean the text data by removing URLs, mentions, hashtags, special characters, and converting to lowercase. Then apply this function to the 'text' column of both dataframes.



```python
import re

def clean_text(text):
    # Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    # Remove mentions
    text = re.sub(r'@\w+', '', text)
    # Remove hashtags
    text = re.sub(r'#\w+', '', text)
    # Remove special characters and keep only alphanumeric and spaces
    text = re.sub(r'[^A-Za-z0-9 ]+', '', text)
    # Convert to lowercase
    text = text.lower()
    return text

train_df['cleaned_text'] = train_df['text'].apply(clean_text)
test_df['cleaned_text'] = test_df['text'].apply(clean_text)

display(train_df[['text', 'cleaned_text']].head())
display(test_df[['text', 'cleaned_text']].head())
```

**Reasoning**:
The text cleaning function has been applied to create the 'cleaned_text' column. The next step is to optionally tokenize the cleaned text.



**Reasoning**:
The previous command failed because the 'punkt_tab' resource was not found. The error message suggests downloading it. I will download 'punkt_tab' and then attempt to tokenize the text again.



```python
import nltk
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize

train_df['tokenized_text'] = train_df['cleaned_text'].apply(word_tokenize)
test_df['tokenized_text'] = test_df['cleaned_text'].apply(word_tokenize)

display(train_df[['cleaned_text', 'tokenized_text']].head())
display(test_df[['cleaned_text', 'tokenized_text']].head())
```

## Exploratory data analysis (eda) on text (future step)

### Subtask:
Analyze word frequencies, tweet lengths, and potentially visualize word clouds for both disaster and non-disaster tweets to gain insights into the language patterns.


**Reasoning**:
Calculate the length of the cleaned text for each tweet and store it in a new column.



```python
train_df['tweet_length'] = train_df['cleaned_text'].apply(len)
test_df['tweet_length'] = test_df['cleaned_text'].apply(len)

display(train_df[['cleaned_text', 'tweet_length']].head())
display(test_df[['cleaned_text', 'tweet_length']].head())
```

**Reasoning**:
Separate the training data into disaster and non-disaster tweets based on the 'target' column.



```python
disaster_tweets = train_df[train_df['target'] == 1]
non_disaster_tweets = train_df[train_df['target'] == 0]

print("Disaster Tweets:")
display(disaster_tweets.head())

print("\nNon-Disaster Tweets:")
display(non_disaster_tweets.head())
```

**Reasoning**:
Determine the most frequent words in disaster and non-disaster tweets using the tokenized text and Counter, and then visualize the top words using bar plots to compare word frequencies.



```python
from collections import Counter
import matplotlib.pyplot as plt

# Flatten the list of tokenized words
disaster_words = [word for tokens in disaster_tweets['tokenized_text'] for word in tokens]
non_disaster_words = [word for tokens in non_disaster_tweets['tokenized_text'] for word in tokens]

# Get the most common words
disaster_word_counts = Counter(disaster_words)
non_disaster_word_counts = Counter(non_disaster_words)

top_n = 20

top_disaster_words = disaster_word_counts.most_common(top_n)
top_non_disaster_words = non_disaster_word_counts.most_common(top_n)

# Visualize the top words
fig, axes = plt.subplots(1, 2, figsize=(15, 7))

# Disaster tweets
words, counts = zip(*top_disaster_words)
axes[0].bar(words, counts, color='skyblue')
axes[0].set_title('Top {} Most Frequent Words in Disaster Tweets'.format(top_n))
axes[0].set_xlabel('Words')
axes[0].set_ylabel('Frequency')
axes[0].tick_params(axis='x', rotation=45)

# Non-disaster tweets
words, counts = zip(*top_non_disaster_words)
axes[1].bar(words, counts, color='lightcoral')
axes[1].set_title('Top {} Most Frequent Words in Non-Disaster Tweets'.format(top_n))
axes[1].set_xlabel('Words')
axes[1].set_ylabel('Frequency')
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

**Reasoning**:
Analyze the distribution of tweet lengths for both disaster and non-disaster tweets by plotting histograms of the 'tweet_length' column for each group.



```python
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Disaster tweets
axes[0].hist(disaster_tweets['tweet_length'], bins=50, color='skyblue', edgecolor='black')
axes[0].set_title('Distribution of Tweet Lengths in Disaster Tweets')
axes[0].set_xlabel('Tweet Length')
axes[0].set_ylabel('Frequency')

# Non-disaster tweets
axes[1].hist(non_disaster_tweets['tweet_length'], bins=50, color='lightcoral', edgecolor='black')
axes[1].set_title('Distribution of Tweet Lengths in Non-Disaster Tweets')
axes[1].set_xlabel('Tweet Length')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

## Feature engineering (future step)

### Subtask:
Convert the text data into numerical features that can be used by machine learning models. This could involve techniques like Bag-of-Words, TF-IDF, or using pre-trained word embeddings.


**Reasoning**:
Import the necessary library and convert the cleaned text into numerical features using TF-IDF.



```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer(stop_words='english', max_features=5000)

X_train_tfidf = tfidf_vectorizer.fit_transform(train_df['cleaned_text'])
X_test_tfidf = tfidf_vectorizer.transform(test_df['cleaned_text'])

print("Shape of X_train_tfidf:", X_train_tfidf.shape)
print("Shape of X_test_tfidf:", X_test_tfidf.shape)
```

## Model selection and training (future step)

### Subtask:
Choose and train a classification model suitable for text data (e.g., Naive Bayes, Logistic Regression, or a simple deep learning model like a Recurrent Neural Network or Transformer).


**Reasoning**:
Import Logistic Regression model and train it using the TF-IDF transformed training data and the target variable.



```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train_tfidf, train_df['target'])
```

## Model evaluation (future step)

### Subtask:
Evaluate the performance of the trained model using appropriate metrics for classification (e.g., accuracy, precision, recall, F1-score).


**Reasoning**:
Import necessary metrics from sklearn.metrics, make predictions on the training data, calculate evaluation metrics, and print them.



```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

y_train_pred = model.predict(X_train_tfidf)

accuracy = accuracy_score(train_df['target'], y_train_pred)
precision = precision_score(train_df['target'], y_train_pred)
recall = recall_score(train_df['target'], y_train_pred)
f1 = f1_score(train_df['target'], y_train_pred)

print(f"Training Accuracy: {accuracy:.4f}")
print(f"Training Precision: {precision:.4f}")
print(f"Training Recall: {recall:.4f}")
print(f"Training F1 Score: {f1:.4f}")
```

## Prediction and submission (future step)

### Subtask:
Make predictions on the test set and format the predictions for submission to the Kaggle competition.


**Reasoning**:
Use the trained Logistic Regression model to predict the target on the test set, then create a submission DataFrame and save it to a CSV file.



```python
test_predictions = model.predict(X_test_tfidf)

submission_df = pd.DataFrame({'id': test_df['id'], 'target': test_predictions})

submission_df.to_csv('submission.csv', index=False)

display(submission_df.head())
```
